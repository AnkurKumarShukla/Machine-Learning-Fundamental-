```python
def changer(x):
  return np.reshape(x,(28,50))
```

# function to visualise the data 

```python
def ankmagic(x):
  xfinal=[0]*len(x)
  for i in range (len(x)):
    xfinal[i]= changer(x[i])/255
   
  return xfinal
```

```python
from sklearn.model_selection import train_test_split
X_train, X_cv, y_train, y_cv = train_test_split(xtrain, ytrain, 
                                                test_size = 0.2,
                                                random_state = 1212)

# X_train = X_train.as_matrix().reshape(33600, 784) #(33600, 784)
# X_cv = X_cv.as_matrix().reshape(8400, 784) #(8400, 784)

# X_test = X_test.as_matrix().reshape(28000, 784)
```

```python
X_train = np.array(X_train,dtype=np.float32); X_cv= np.array(X_cv,dtype=np.float32); X_test =np.array(X_test,dtype=np.float32)
# data = np.array(data, dtype=np.float32)
# normalising the image data for better training
X_train /= 255; X_cv /= 255; X_test /= 255

# Convert labels to One Hot Encoded to feed into the model
num_digits = 20
y_train = keras.utils.np_utils.to_categorical(y_train, num_digits)
y_cv = keras.utils.np_utils.to_categorical(y_cv, num_digits)
```

```python
print(y_train[0]) # 2
print(y_train[3]) # 7
```

```python
n_input = 1400 # number of features
n_hidden_1 = 300 # 1st layer number of features
n_hidden_2 = 100
n_hidden_3 = 100
n_hidden_4 = 200
num_digits = 20  # 20 digits recognition
```

```python
from keras.layers import Input  # input layer import to make the input layer
```

```python
import pandas as pd
import numpy as np

np.random.seed(1212)

import keras
from keras.models import Model
from keras.layers import *
from keras import optimizers
```

```python

#  using keras sequential model simple neural network model but we can also use convolutional model

Inp = Input(shape=(1400,))
x = Dense(n_hidden_1, activation='relu', name = "Hidden_Layer_1")(Inp)
x = Dense(n_hidden_2, activation='relu', name = "Hidden_Layer_2")(x)
x = Dense(n_hidden_3, activation='relu', name = "Hidden_Layer_3")(x)
x = Dense(n_hidden_4, activation='relu', name = "Hidden_Layer_4")(x)
output = Dense(num_digits, activation='softmax', name = "Output_Layer")(x)
```

```python
# Our model would have '6' layers - input layer, 4 hidden layer and 1 output layer
model = Model(Inp, output)
model.summary() 
```
